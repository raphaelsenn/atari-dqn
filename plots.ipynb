```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

## Breakout

```python
file_name = "DQN-ALE-Breakout-v5-lr0.00025-report.csv"
data = pd.read_csv(file_name)
data.head()
```

```python
t = data["Timestep"] / 50_000

columns = [f"score_{i}" for i in range(5)]
scores = data[columns].to_numpy()
scores.shape
```

```python
mean = scores.mean(axis=1)

fig, ax = plt.subplots()
ax.plot(t[:200], mean[:200], color="blue", alpha=0.7)
ax.set_xlabel("Training Epochs")
ax.set_ylabel("Average Score per Episode")
plt.show()
```
